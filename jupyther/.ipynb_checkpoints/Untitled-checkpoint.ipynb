# Linear Regression
Linear regression the expected value of $y$ is assumed to be a linear combination of the input variables $x_j$. In mathematical notion, if $\hat{y}$ is the predicted value.
$\hat{y}(\boldsymbol \beta, \mathbf x) = \beta_0 + \beta_1 x_1 + ... + \beta_p x_p$
The vector $\boldsymbol \beta$ and $\beta_0$ as `intercept`.
Later we perform classification with generalized linear models, using logistic regression.

# Simulating a cure and running linear regression
We need to start with a defining a curve and add some noise about this curve. Use the exponential function for instance $y=e^{3x}$.


```r
x_tr <- seq(0, 2, length=200)
y_tr <- exp(3*x_tr)
```

```r
plot(x_tr, y_tr, type="l")
```

```r
set.seed(10)
y <- y_tr+rnorm(length(x_tr), mean=0, sd=25)
```

```r
plot(x_tr,y)
```

```r
plot(x_tr,y)
points(x_tr, y_tr, type="l", col="red")
```

# Fit a line
Now let's fit a line, this is a wrong model, but can be regarded as an approximation.

```r
lr.fit <- lm(y_tr~x_tr)
yhat <- predict(lr.fit)
```

```r
plot(x_tr,y)
points(x_tr, y_tr, type="l", col="red")

points(x_tr, yhat,type="l",col="green")
```

# Binary regression
Now lets make a binary response variable. In most supervised learning problems is category is to be learned. You can look at binary data as a continuous data out of a threshold somewhere.

Lets threshold this data at point zero. 





```r
z=(1*(y>90)-0.5)*2
```

```r
z
```

```r
plot(x_tr, z)
```

# Exercise
* fit a linear regression on z as the response (and x_tr as the feature)

```r
zhat <- predict(lm(z~x_tr))
```

```r
zclass= 2*(zhat>0)-1
```

```r
mean(zclass==z)
```

```r
plot(x_tr,z, col= z+3)
points(x_tr, zhat, type="l")
points(x_tr, zclass, type="l", col="green")
```

```r
require(MASS)
n <- 50
X1 <- mvrnorm(n, mu=c(1,1),Sigma=diag(c(1,1)))
X2 <- mvrnorm(n, mu=c(-1,-1),Sigma=diag(c(1,1)))
X<-rbind(X1,X2)
z <- rep(c(-1,1), each=n)
              
              
            
```

```r
plot(X[,1], X[,2],col=z+3)
```

```r
lm.obj<- lm(z~X[,1]+X[,2])
zhat <- predict(lm.obj)
zhat
```
